```python
import tensorflow as tf
```

```python
x = tf.random.normal([10])
x.numpy()
```

```python
tf.gather(x,[0,2]).numpy()
```

```python
x = tf.random.normal([10,10])
x.numpy()
```

```python
tf.gather(x,[0,1,2]).numpy()
```

```python
x = tf.random.normal([128,5,2])
x
```

```python
tf.gather(x,[0,23],batch_dims=0,axis=1)
```

```python
tf.gather(x,[0,1],axis=1,batch_dims=0).shape
```

```python
tf.gather(x[0:5],[0,0,0,0,0],axis=1,batch_dims=0)[:,0,:]
```
