```python
x = tf.random.normal(shape=[10])
x
```

```python
x = tf.random.normal(shape=(128,5,2))
probs = tf.random.normal(shape=(128,5,5))
```

```python
total_rank = tf.argsort(probs, axis=2, direction="DESCENDING")
total_rank
```

```python
idx = total_rank[:, :, 1]
idx
```

```python
pointer = tf.gather(x, idx, batch_dims=0)
pointer
```
