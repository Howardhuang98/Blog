```python
indices = tf.constant([[0], [2]])
updates = tf.constant([[[5, 5, 5, 5], [6, 6, 6, 6],
                            [7, 7, 7, 7], [8, 8, 8, 8]],
                           [[5, 5, 5, 5], [6, 6, 6, 6],
                            [7, 7, 7, 7], [8, 8, 8, 8]]])
shape = tf.constant([4, 4, 4])
scatter = tf.scatter_nd(indices, updates, shape)
print(scatter)
print(updates.shape)
print(indices.shape)
```

对于一个全零的 [4,4,4]
它将[0,:,:] 和 [2,:,:]做了更新
所以说：
indice 是 [[位置1]，[位置2]，...[位置n]]
update 是 [[位置1对应的更新]，[位置2对应的更新]，...[位置n对应的更新]]

```python
indices = tf.constant([[0,1], [2,1]])
updates = tf.constant([[5, 5, 5, 5],
                           [8, 8, 8, 8]])
shape = tf.constant([4, 4, 4])
scatter = tf.scatter_nd(indices, updates, shape)
print(scatter)
print(updates.shape)
print(indices.shape)
```

```python
idx = tf.ones(shape=(200))
step = tf.constant(0)
pointed = tf.constant(-1,shape=(5,200))
tf.tensor_scatter_nd_update(pointed,[[step]],[idx])
```
